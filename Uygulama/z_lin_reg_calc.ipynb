```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.array([2, 1, 3])  # actual params
X = np.array([  # toy samples
    [1, 2, 3],
    [1, 3, 2],
    [1, 4, 5],
    [1, 6, 7]
])
y = np.dot(X, t)  # dependent variable
t_model = np.array([2, 0, 4])  # model's initial random params
losses = []  # store loss at each gradient descent step
model_params = []  # monitor how model's params change over time
lr = 0.05  # learning rate
for i in range(500):
    y_hat = np.dot(X, t_model)  # get model's predictions
    loss = np.sum((y_hat - y) ** 2) / 8  # calculate loss
    losses.append(loss)
    model_params.append(t_model)
    d_t = np.dot(X.T, (y_hat - y)) / 4  # calculate gradients
    t_model = t_model - lr * d_t  # update model params
    if (i+1) % 20 == 0:  # log the progress
        print(i, "\tJ:", np.round(loss, 4), "\tM:", np.round(t_model, 4))

losses = np.array(losses)
model_params = np.array(model_params)
```

```python
plt.title("Loss")
plt.plot(losses)
plt.show()
```

```python
plt.title("Theta 0")
plt.plot((model_params.T)[0])
plt.show()
```

```python
plt.title("Theta 1")
plt.plot((model_params.T)[1])
plt.show()
```

```python
plt.title("Theta 2")
plt.plot((model_params.T)[2])
plt.show()
```
